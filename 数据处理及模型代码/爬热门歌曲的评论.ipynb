```python
result=[]
with open('ip.txt',encoding='UTF-8') as f:
    for line in f:
        result.append(list(line.strip('\n').split(',')))

def get_proxies():
    "https://" +random.choice(result)[0]
    headers = {'https:':('https://'+random.choice(result)[0])} 
    return headers
```

```python
#åˆ›å»ºå‡½æ•°ï¼Œè·å–è¯„è®ºä¿¡æ¯
def get_comments(page):
    """
    è·å–è¯„è®ºä¿¡æ¯
    """
    url = 'http://music.163.com/api/v1/resource/comments/R_SO_4_1411358329?limit=20&offset=' + str(page)
    response = requests.get(url=url, headers=headers,proxies = get_proxies())    
    result = json.loads(response.text)# å°†å­—ç¬¦ä¸²è½¬ä¸ºjsonæ ¼å¼
    items = result['comments']
    data={}
    for item in items:       
        data['user_name'] = item['user']['nickname'].replace(',', 'ï¼Œ')#ç”¨æˆ·å
        data['user_id'] = str(item['user']['userId']) # ç”¨æˆ·ID
        data['comment'] = item['content'].strip().replace('\n', '').replace(',', 'ï¼Œ')# è¯„è®ºå†…å®¹    
        date = time.localtime(int(str(item['time'])[:10]))# è¯„è®ºæ—¶é—´
        data['date'] = time.strftime("%Y-%m-%d %H:%M:%S", date)
        print(data)
#         with open('æ­Œæ›²è¯„è®º1.json', 'w') as f:
#         jsObj = json.dumps(data)
#         f.write(jsObj+'\n')
        with open('è¯„è®º/test_å‹‡æ°”.csv', 'a', encoding='utf-8-sig') as f2:
            f2.write( data['user_name']+ ',' + data['user_id'] + ',' + data['comment'] + ',' + data['date'] + '\n')
        f2.close()
#             f.write(data)
        
```

```python
# alldata=[]
# f = open("è¯„è®º/éº»é›€è¯„è®º2.json", "w")
for i in range(0,10000, 20):
    get_comments(i)
# f.close()
```
