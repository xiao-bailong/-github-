```python
#首先导入需求库
import json
import time
import requests
import random
```

```python
#设置headers信息，模拟成浏览器去访问网站，跳过反爬虫机制
headers = { 'Host': 'music.163.com', 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
```

```python
result=[]
with open('ip.txt',encoding='UTF-8') as f:
    for line in f:
        result.append(list(line.strip('\n').split(',')))

def get_proxies():
    "https://" +random.choice(result)[0]
    headers = {'https:':('https://'+random.choice(result)[0])} 
    return headers
```

```python
#创建函数，获取评论信息
def get_comments(page):
    """
    获取评论信息
    """
    url = 'http://music.163.com/api/v1/resource/comments/R_SO_4_1411358329?limit=20&offset=' + str(page)
    response = requests.get(url=url, headers=headers,proxies = get_proxies())    
    result = json.loads(response.text)# 将字符串转为json格式
    items = result['comments']
    data={}
    for item in items:       
        data['user_name'] = item['user']['nickname'].replace(',', '，')#用户名
        data['user_id'] = str(item['user']['userId']) # 用户ID
        data['comment'] = item['content'].strip().replace('\n', '').replace(',', '，')# 评论内容    
        date = time.localtime(int(str(item['time'])[:10]))# 评论时间
        data['date'] = time.strftime("%Y-%m-%d %H:%M:%S", date)
        print(data)
#         with open('歌曲评论1.json', 'w') as f:
#         jsObj = json.dumps(data)
#         f.write(jsObj+'\n')
        with open('评论/test_勇气.csv', 'a', encoding='utf-8-sig') as f2:
            f2.write( data['user_name']+ ',' + data['user_id'] + ',' + data['comment'] + ',' + data['date'] + '\n')
        f2.close()
#             f.write(data)
        
```

```python
# alldata=[]
# f = open("评论/麻雀评论2.json", "w")
for i in range(0,10000, 20):
    get_comments(i)
# f.close()
```
