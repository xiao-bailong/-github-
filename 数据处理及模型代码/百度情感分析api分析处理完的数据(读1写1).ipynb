```python
dataset=pd.read_csv("评论/去非中英文符号/test_勇气.csv",encoding='utf-8-sig',header=None)
dataset.columns = ['name', 'id', 'comment', 'time']
# dataset.columns = ['name', 'id', 'comment', 'time','positive_prob','sentiment']
col_name = dataset.columns.tolist()
col_name.insert(4,'positive_prob')
col_name.insert(5,'sentiment')
dataset=dataset.reindex(columns=col_name)
dataset
```

```python
x=dataset['comment']
#0消极，1中肯，2积极
for i in range(len(x)):
    time.sleep(0.2)
    text=x[i]
    # 通过百度提供的接口方法进行情感倾向提取
    result=client.sentimentClassify(text)#去了;
    print(i)
    print(result)
    # 如果解析错误则填写上空值,使得程序不会出错而停止运行
    if "error_code" in result.keys():
        dataset['positive_prob'][i] = " "
        dataset['sentiment'][i] = " "
    else:
        data = result['items']
        items = data[0]
        dataset['positive_prob'][i] = items['positive_prob']
        dataset['sentiment'][i] = items['sentiment']
        
dataset.to_csv('评论/处理完/test_勇气.csv', index=False,encoding='utf-8-sig')
```
